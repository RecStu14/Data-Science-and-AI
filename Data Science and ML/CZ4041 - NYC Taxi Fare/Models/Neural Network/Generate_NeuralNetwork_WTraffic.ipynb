# Running Neural Networks with Traffic

```python
import tensorflow as tf
import pandas as pd
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
checkpoint5_traffic = pd.read_csv('../checkpoint5_traffic.csv')
checkpoint5_traffic
```

```python
labels = checkpoint5_traffic[['fare_amount']]
labels
```

```python
del checkpoint5_traffic['fare_amount']
checkpoint5_traffic
```

```python
del checkpoint5_traffic['time']
checkpoint5_traffic
```

```python
sc = StandardScaler()
checkpoint_5_scaled = np.array(sc.fit_transform(checkpoint5_traffic))
labels_scaled = np.array(labels)
print("successfully scaled...")
```

```python
print("splitting...")
#splitting data into train test 
X_train, X_test, y_train, y_test = train_test_split(checkpoint_5_scaled, labels_scaled, test_size=0.2, random_state=42)
```

```python
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

## Creating the Neural Network

```python
#Creating a Simple Neural Network model
model = Sequential()
#The input_shape is kept at 12 as we have 12 features that we are sending into the neural network
model.add(Dense(16, input_shape=(12,)))
model.add(Dense(16, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='linear'))

model.compile(optimizer='adam', loss='mse', metrics=[tf.keras.metrics.RootMeanSquaredError(),tf.keras.metrics.MeanAbsoluteError()])
```

```python
print(model.summary())
```

```python
history = model.fit(X_train, y_train, epochs = 30)

model.save('NN_traffic_density_v2.h5')

history_df = pd.DataFrame(history.history)
history_df.to_csv('NN_traffic_History.csv',index=False )
```

```python
#Evaluating on X_test
model.evaluate(X_test, y_test)
```
