# CREATING LABELS

Creating the labels for traffic using equal weightage for the weights of the 3 terms in the equation. This converts the unsupervised problem into a supervised problem. Equation:

#### Traffic = A·max((weekend || holiday), 0.05 ) + B·max(peak_hr,0.1) + C·max(hotspot, 0.1)+ ɛ

```python
#function to calculate the equation
def calc_traffic_density(weekend, public_hol, peak_hr, hotspot):
    #defining the weights and the small constant
    WEIGHT_A = WEIGHT_B = WEIGHT_C = 1/3
    EPSILON = sys.float_info.epsilon
    
    #Getting the Unweighted values
    day = max((weekend or public_hol),0.05)
    time = max(peak_hr,0.1)
    place = max(hotspot,0.1)
    
    #Getting the Weighted Values
    weighted_day = WEIGHT_A * day
    weighted_time = WEIGHT_B * time
    weighted_place = WEIGHT_C * place
    
    #Summing up all the variables
    traffic = weighted_day + weighted_place + weighted_time + EPSILON
    
    return traffic
```

```python
#Iterating through the dataset
traffic_density = []
for index, row in checkpoint5.iterrows():
    traffic_value = calc_traffic_density(weekend=row['weekend'], public_hol=row['holiday'], peak_hr=row['peak_hour'], hotspot=row['hotspot'])
    traffic_density.insert(index,traffic_value)
```

```python
#Iterating through the dataset - kaggle test
traffic_density_kt = []
for index, row in kaggle_test.iterrows():
    traffic_value_kt = calc_traffic_density(weekend=row['weekend'], public_hol=row['holiday'], peak_hr=row['peak_hour'], hotspot=row['hotspot'])
    traffic_density_kt.insert(index,traffic_value_kt)
```
